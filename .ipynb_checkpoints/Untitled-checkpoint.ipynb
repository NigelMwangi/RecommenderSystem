## Modelling

Here we will involve building and training recommendation algorithms to provide personalized movie suggestions to users. We will create collaborative filtering recommendation model using Surprise library, which is a popular library for building recommender systems.

We will load the data again and perform Data Splitting, training and testing sets that will evaluate model performance.

```python
# Loading and Preparing Data

data = pd.read_csv('merged_ratings_movies.csv')

# Define the reader for Surprise
reader = Reader(rating_scale=(1, 5))

# Load data into Surprise's Dataset
data = Dataset.load_from_df(data[['userId', 'movieId', 'rating']], reader)

# Split the data into training and testing sets
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
```

Here we will choose a collaborative filtering algorithm. We'll use Singular Value Decomposition (SVD), as our Model selection

```python
# Here we Build a recommendation model  and train it using the training set:
model = SVD()
model.fit(trainset)
```

We then do Model Evaluation, thiis is to evaluate the models performance using the MAE(Mean Absolute Error) and RMSE(Root Mean Squared Error) metrics

```python
# Make predictions on the test set
predictions = model.test(testset)

# Calculate RMSE (Root Mean Squared Error)
rmse = accuracy.rmse(predictions)

# You can also calculate other metrics like MAE
mae = accuracy.mae(predictions)
```

```python
# Performed cross-validation with 5 k models
from surprise.model_selection import cross_validate
results = cross_validate(model, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)


print("Cross-Validation Results:")
for key, value in results.items():
    print(f'{key}: {value}')
```

Now, lets get a<b> preview </b>of our Users movie Titles that have been recommended to:

```python
def print_top_n_recommendations(ranked_predictions, merged_data, n):

    if n > len(ranked_predictions):
        print("Requested number of recommendations exceeds available recommendations.")
        n = len(ranked_predictions)
    
   
    print(f"Top {n} Recommended Movies:")
    for i in range(n):
        movie_id, predicted_rating = ranked_predictions[i]
        
        
        title = merged_data[merged_data['movieId'] == movie_id]['title'].values[0]
        
        print(f"{i + 1}. Movie Title: {title}, Predicted Rating: {predicted_rating:.2f}")

# Our Preview
print_top_n_recommendations(ranked_predictions, merged_data, 5)
```
